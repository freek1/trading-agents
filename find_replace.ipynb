```python
import os

def find_and_replace_files(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            # Rename the file
            new_file_name = file.replace('-True', '')
            new_file_path = os.path.join(root, new_file_name)
            os.rename(file_path, new_file_path)
            print(f"Renamed file: {new_file_path}")

# Replace 'directory_path' with the actual path to the directory you want to search
directory_path = 'outputs/20230619_192815/'
find_and_replace_files(directory_path)
```
